```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Criação de Data Frame

```python
df_faturamento = dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento
```

```python
df_medidas = pd.DataFrame.from_dict(df_faturamento)
```

```python
df_medidas
```

```python
#Variância
df_medidas.valor.var()
```

```python
#Desvio Padrão
df_medidas.valor.std
```

```python
#Calculo de assimetria
df_medidas.valor.skew()
```

```python
#Calculo de curtose
df_medidas.valor.kurtosis()
```

```python
#Diagrams vertical para verificar o quanto de valores foram arrecados por ano.
df_medidas.plot.bar(x='data_ref', y='valor')
```

```python
#Nesse grafico de barra com o metodo .sort ele demonstra de ordem crescente pelas datas os valores.
df_medidas.sort_values('valor').plot.bar(x='data_ref', y='valor')
```

```python
#Nesse grafico vai demonstrar barras horizontais, de ordem de valores do menor para o maior.
df_medidas.sort_values('valor').plot.barh(x='data_ref', y='valor')
```

```python
#grafico de linhas para demonstrar por mes valor e a referência
df_medidas.plot.line(x='data_ref', y='valor')
```

```python
#Calculo da média de vendas
media = df_medidas['valor'].mean
```
